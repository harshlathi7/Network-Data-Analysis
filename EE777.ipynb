```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("EE777_Data.csv")
```

You'll need to execute the above cell after each and every execution of the below ones. 

```
df = df.pivot(index='Time', columns='City', values='RTT')

df.plot().set(xlabel='Time',ylabel='RTT(ms)')
plt.savefig('RTTvsTimePerCity.png')
```

```
df = df.pivot(index='Time', columns='City', values='Jitter')

df.plot().set(xlabel='Time',ylabel='Jitter(ms)')
plt.savefig('JittervsTimePerCity.png')
```

```
df = df.pivot(index='Time', columns='City', values='Packets lost')

df.plot().set(xlabel='Time',ylabel='Packets lost')
plt.savefig('PacketsLostvsTimePerCity.png')
```

```
df = df.pivot(index='Time', columns='City', values='Loss %')

df.plot().set(xlabel='Time',ylabel='Loss percentage')
plt.savefig('LossPercentagevsTimePerCity.png')
```

```
df = df.pivot(index='Time', columns='City', values='Download')

df.plot().set(xlabel='Time',ylabel='Download Speed(Mbps)')
plt.savefig('DownloadvsTimePerCity.png')
```

```
df = df.pivot(index='Time', columns='City', values='Upload')

df.plot().set(xlabel='Time',ylabel='Upload Speed(Mbps)')
plt.savefig('UploadvsTimePerCity.png')
```
